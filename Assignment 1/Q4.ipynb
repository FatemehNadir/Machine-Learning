#### we calculate sigma with the fisher information that is the negative expected value of this second derivative:


```python
import numpy as np
from   scipy.stats import norm
import matplotlib.pyplot as plt


# Plot the asymptotically normal distribution.
N  = 11
p0 = 0.6
xx = np.arange(0, 1, 0.001)
yy = norm.pdf(xx, p0, np.sqrt((p0 * (1-p0)) / N))

# Generate many random samples of size N and compute MLE.
mles = []
for _ in range(N*100):
    X = np.random.binomial(1, p0, size=N)
    mles.append(X.mean())

# Plot histogram of MLEs.
plt.plot(xx, yy)
plt.hist(mles, bins=20, density=True)
plt.savefig('as1.png')
plt.show()
```
